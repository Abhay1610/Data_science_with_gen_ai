## Question 1: What is a Decision Tree, and how does it work in the context of classification?


A decision tree is a supervised learning algorithm, used for both classification and regression tasks, that models decisions in a tree-like structure. In classification, it works by recursively partitioning the data based on feature values, leading to a prediction at the leaf nodes. Essentially, it's like a flowchart that helps classify data by asking a series of questions.

Here's how it works in classification:

Structure:
1. A decision tree consists of a root node, internal nodes, and leaf nodes.
2. The root node represents the initial decision based on the most important feature.
3. Internal nodes: represent further decisions based on other features, branching out to different subsets of the data.
4. Leaf nodes: represent the final classifications or predictions for the data points that reach them.

## Question 2: Explain the concepts of Gini Impurity and Entropy as impurity measures. How do they impact the splits in a Decision Tree?


Decision Trees use Gini Impurity and Entropy to determine the best way to split the data at each node:

1. Gini Impurity: Measures the probability of misclassifying a randomly chosen element based on the class distribution within a node. Lower Gini impurity signifies a purer, more homogeneous node.

2. Entropy: Quantifies the randomness or disorder within a node's data points regarding classification. Lower entropy means less uncertainty and a more pure node.

Impact on Splits:

a) Decision tree algorithms like CART (for Gini) and ID3/C4.5 (for Entropy) aim to find the splits that result in the largest reduction in impurity (Gini Impurity or Entropy).

b) The split that minimizes impurity (maximizes purity) is selected, creating more homogeneous child nodes and ultimately improving the accuracy of the decision tree.

c) Gini Impurity is generally faster to compute as it doesn't involve logarithmic calculations, while Entropy can provide slightly better results in some cases, especially with imbalanced datasets. The choice often depends on the specific needs of the problem and the dataset characteristics.


## Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision Trees? Give one practical advantage of using each.

Decision trees are powerful and intuitive machine learning models, but they are susceptible to overfitting the training data, leading to poor generalization on unseen examples. To address this, two main pruning techniques are employed: pre-pruning and post-pruning.

1. Pre-pruning (early stopping): Stop growing the decision tree during its construction based on pre-defined criteria (e.g., maximum depth, minimum samples per leaf).

 Advantage: Faster training due to smaller trees.

2. Post-pruning: Build the complete decision tree and then remove branches or nodes that don't improve the model's accuracy on a separate validation set.

 Advantage: Better generalization, as it considers all patterns before simplification

## Question 4: What is Information Gain in Decision Trees, and why is it important for choosing the best split?


Information Gain (IG) in Decision Trees measures how much a feature improves the "purity" or homogeneity of data subsets after splitting based on that feature.



*   Purity (or Homogeneity): Refers to how much the data within a subset belongs to the same class. A perfectly pure subset has all data points belonging to a single class, resulting in lower uncertainty.
*   How it works: IG compares the entropy (a measure of uncertainty) of the dataset before and after a split. The greater the reduction in entropy, the higher the information gain for that split.
*   Importance: Features with the highest IG are chosen to split nodes in the decision tree, leading to a more efficient and accurate classification process by prioritizing the most informative features. In simple terms, it helps decide the best questions to ask at each stage of the tree to create clearer, more manageable subsets of data.

### Question 6: Write a Python program to:
 Load the Iris Dataset

 Train a Decision Tree Classifier using the Gini criterion

 Print the models accuracy and feature importances

```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train Decision Tree Classifier using Gini criterion
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(X_train, y_train)

# Predict on test data
y_pred = clf.predict(X_test)

# Print model accuracy
print("Model Accuracy:", accuracy_score(y_test, y_pred))

# Print feature importances
print("Feature Importances:")
for name, importance in zip(iris.feature_names, clf.feature_importances_):
    print(f"{name}: {importance:.4f}")

# Visualize the decision tree
plt.figure(figsize=(12, 8))
plot_tree(
    clf,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True
)
plt.show()
```

### Question 7: Write a Python program to:
 Load the Iris Dataset

 Train a Decision Tree Classifier with max_depth=3 and compare its accuracy to
a fully-grown tree.


```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train Decision Tree with max_depth=3
clf_depth3 = DecisionTreeClassifier(max_depth=3, random_state=42)
clf_depth3.fit(X_train, y_train)
y_pred_depth3 = clf_depth3.predict(X_test)

# Train fully grown Decision Tree (no depth limit)
clf_full = DecisionTreeClassifier(random_state=42)
clf_full.fit(X_train, y_train)
y_pred_full = clf_full.predict(X_test)

# Compare accuracies
print("Decision Tree (max_depth=3) Accuracy:", accuracy_score(y_test, y_pred_depth3))
print("Decision Tree (fully grown) Accuracy:", accuracy_score(y_test, y_pred_full))

# Plot both trees side by side
plt.figure(figsize=(20, 8))

# Tree with max_depth=3
plt.subplot(1, 2, 1)
plot_tree(clf_depth3,
          feature_names=iris.feature_names,
          class_names=iris.target_names,
          filled=True, rounded=True)
plt.title("Decision Tree (max_depth=3)")

# Fully grown tree
plt.subplot(1, 2, 2)
plot_tree(clf_full,
          feature_names=iris.feature_names,
          class_names=iris.target_names,
          filled=True, rounded=True)
plt.title("Decision Tree (Fully Grown)")

plt.show()
```

### Question 9: Write a Python program to:
 Load the Iris Dataset

 Tune the Decision Trees max_depth and min_samples_split using
GridSearchCV

 Print the best parameters and the resulting model accuracy

```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Define parameter grid for tuning
param_grid = {
    "max_depth": [2, 3, 4, 5, None],
    "min_samples_split": [2, 3, 4, 5, 6, 10]
}

# Initialize Decision Tree and GridSearchCV
dt = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=dt,
    param_grid=param_grid,
    cv=5,            # 5-fold cross-validation
    scoring="accuracy",
    n_jobs=-1
)

# Fit the model
grid_search.fit(X_train, y_train)

# Get the best parameters
print("Best Parameters:", grid_search.best_params_)

# Evaluate on test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print("Model Accuracy on Test Set:", accuracy_score(y_test, y_pred))
```

 Handle the missing values

 Encode the categorical features

 Train a Decision Tree model

 Tune its hyperparameters

 Evaluate its performance

And describe what business value this model could provide in the real-world
setting.


**Answer**:
###  Step-by-Step Process
**1. Handle Missing Values**

* Check % of missing values per feature.
* If few  drop rows; if many  impute (mean/median for numeric, mode/frequent for categorical).
* Consider advanced imputation (KNN/Iterative Imputer) for better accuracy.

**2. Encode Categorical Features**

* Use **One-Hot Encoding** for nominal (no order) variables.
* Use **Ordinal Encoding** if categories have a natural order.
* Ensure encoding is consistent across train/test splits.

**3. Train Decision Tree Model**

* Split data into **train & test sets**.
* Train a **DecisionTreeClassifier** on processed data.

**4. Tune Hyperparameters**

* Use **GridSearchCV / RandomizedSearchCV** to optimize:

  * `max_depth` (controls tree depth, prevents overfitting)
  * `min_samples_split` & `min_samples_leaf`
  * `criterion` (e.g., Gini vs Entropy)

**5. Evaluate Performance**

* Use **Accuracy, Precision, Recall, F1-score**.
* Plot **Confusion Matrix** to see correct vs wrong predictions.
* Consider **ROC-AUC** for imbalanced datasets.
